```python
from sklearn import preprocessing
from sklearnex import patch_sklearn
patch_sklearn()
```

```python
le=preprocessing.LabelEncoder()
df['GENDER']=le.fit_transform(df['GENDER'])
df['LUNG_CANCER']=le.fit_transform(df['LUNG_CANCER'])
df['SMOKING']=le.fit_transform(df['SMOKING'])
df['YELLOW_FINGERS']=le.fit_transform(df['YELLOW_FINGERS'])
df['ANXIETY']=le.fit_transform(df['ANXIETY'])
df['PEER_PRESSURE']=le.fit_transform(df['PEER_PRESSURE'])
df['CHRONIC DISEASE']=le.fit_transform(df['CHRONIC DISEASE'])
df['FATIGUE ']=le.fit_transform(df['FATIGUE '])
df['ALLERGY ']=le.fit_transform(df['ALLERGY '])
df['WHEEZING']=le.fit_transform(df['WHEEZING'])
df['ALCOHOL CONSUMING']=le.fit_transform(df['ALCOHOL CONSUMING'])
df['COUGHING']=le.fit_transform(df['COUGHING'])
df['SHORTNESS OF BREATH']=le.fit_transform(df['SHORTNESS OF BREATH'])
df['SWALLOWING DIFFICULTY']=le.fit_transform(df['SWALLOWING DIFFICULTY'])
df['CHEST PAIN']=le.fit_transform(df['CHEST PAIN'])
df['LUNG_CANCER']=le.fit_transform(df['LUNG_CANCER'])
```

```python
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
#Splitting independent and dependent variables
X = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

```python
X
```

```python
y
```

```python
from imblearn.over_sampling import ADASYN
from sklearnex import patch_sklearn
patch_sklearn()
adasyn = ADASYN(random_state=42)
X, y = adasyn.fit_resample(X, y)
```

```python
#Splitting data for training and testing
from sklearn.model_selection import train_test_split
from sklearnex import patch_sklearn
patch_sklearn()
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearnex import patch_sklearn
patch_sklearn()
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
#Model accuracy
from sklearn.metrics import classification_report, accuracy_score, f1_score
from sklearnex import patch_sklearn
patch_sklearn()
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)

score_lr = round(accuracy_score(y_lr_pred,y_test)*100,2)
print("The accuracy of Logistic Regression is: "+str(score_lr)+" %")
```

```python
#Fitting training data to the model
from sklearn.tree import DecisionTreeClassifier
from sklearnex import patch_sklearn
patch_sklearn()
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=0)  
dt_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
#Model accuracy
dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)

score_dt = round(accuracy_score(y_dt_pred,y_test)*100,2)
print("The accuracy of Decision Tree Classifier is: "+str(score_dt)+" %")
```

```python
#Fitting K-NN classifier to the training set  
from sklearn.neighbors import KNeighborsClassifier  
from sklearnex import patch_sklearn
patch_sklearn()
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
knn_model.fit(X_train, y_train)
```

```python
app = Flask(__name__)
```

```python
@ app.route('/')
def home():
    title = 'index'
    return render_template('index.html', title=title)
```

```python
@ app.route('/form')
def form():
    title = 'Form'
    return render_template('form.html', title=title)
```

```python
@ app.route('/result', methods=['GET', 'POST'])
def result():
    if request.method == 'POST':
        gender = int(request.form['gender'])
        age = int(request.form['age'])
        smoking = int(request.form['smoking'])
        yf = int(request.form['yellow-fingers'])
        anx = int(request.form['anxiety'])
        pp = int(request.form['peer-pressure'])
        cd = int(request.form['chronic-disease'])
        fat = int(request.form['fatigue'])
        al = int(request.form['allergy'])
        wh = int(request.form['wheezing'])
        alc = int(request.form['alcohol'])
        cou = int(request.form['coughing'])
        sob = int(request.form['shortness-of-breath'])
        sd = int(request.form['swallowing-difficulty'])
        cp = int(request.form['chest-pain'])
        anxyf = anx*yf
        
        data = [[gender, age, smoking, yf, anx, pp, cd,fat,al,wh,alc,cou,sob,sd,cp,anxyf]]
        accuracy=[98.33,93.33,92.5,95.83,95.83,65.83,95.83,100,98.33]
        l=[]
        
        my_prediction = LR_model.predict(data)
        LR = my_prediction[0]
        l.append(LR)
        my_prediction = DT_model.predict(data)
        DT = my_prediction[0]
        l.append(DT)
        my_prediction = KNN_model.predict(data)
        KNN = my_prediction[0]
        l.append(KNN)
        my_prediction = GNB_model.predict(data)
        GNB = my_prediction[0]
        l.append(GNB)
        my_prediction = MNB_model.predict(data)
        MNB = my_prediction[0]
        l.append(MNB)
        my_prediction = SVC_model.predict(data)
        SVC = my_prediction[0]
        l.append(SVC)
        my_prediction = RF_model.predict(data)
        RF = my_prediction[0]
        l.append(RF)
        my_prediction = MLP_model.predict(data)
        MLP = my_prediction[0]
        l.append(MLP)
        my_prediction = LGB_model.predict(data)
        LGB = my_prediction[0]
        l.append(LGB)
        sp=0
        sn=0
        for i in l:
            if(l[i]==1):
                sp+=accuracy[i]
            else:
                sn+=accuracy[i]
        
        sp=int((l.count(1)/len(l))*100)
        sn=int((l.count(0)/len(l))*100)
        
        if(l.count(1)>l.count(0)):
           return render_template('yes.html',sp=sp)
        else:
           return render_template('no.html',sn=sn)
        
```

```python
if __name__ == '__main__':
    app.run(debug=False,port=9013)
```
